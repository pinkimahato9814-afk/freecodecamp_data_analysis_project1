```python
import pandas as pd
df=pd.read_csv('/content/1994_census_up.csv')
# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Strip whitespace from string data
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df
```

```python
#  How many people of each race are represented in this dataset? This should be a Pandas series with race names as the index labels. (race column)

race_counts = df['race'].value_counts()
print(race_counts)
```

```python
#what is the average age of men
male_avg_age= df[df['sex'] == 'Male']["age"].mean()
print(male_avg_age)
```

```python
#what is the percentage of bachelor degree

bachelor=df[df["education"]=="Bachelors"]
total_bachelor=len(bachelor)
total_person=len(df)
percentage_bachelor=(total_bachelor/total_person)*100
print(f'{round(percentage_bachelor,2)}%')
```

```python
# What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
df_higher_edu=df[df["education"].isin(["Bachelors","Masters","Doctorate"])]
length_higher_degre=len(df_higher_edu[df_higher_edu["salary"]==">50K"])

percentage_earn_more_50k=length_higher_degre/total_person*100
print(f'{round(percentage_earn_more_50k,2)}%')


```

```python
# What percentage of people without advanced education make more than 50K
df_lower_edu  = df["education"].value_counts()
print(df_lower_edu)
```

```python
morethan_50 = df[df["salary"]==">50K"]
print(morethan_50)
```

```python
lower_degree=len(morethan_50)-length_higher_degre
percentage = lower_degree/total_person*100
print(f'{round(percentage,2)}%')
```

```python
# What is the minimum number of hours a person works per week?
min_hours=df["hours-per-week"].min()
print(min_hours)
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
min_max_earning=morethan_50[morethan_50["hours-per-week"]==1]


percentage_min_work_mas_earn=len(min_max_earning)/total_person*100
print(f'{round(percentage_min_work_mas_earn,2)}%')
```

```python
# What country has the highest percentage of people that earn >50K and what is that percentage?
# Identify the most popular occupation for those who earn >50K in India.
```

```python
total_country=morethan_50["native-country"].value_counts()   #value_counts method give  dictionary
# print(total_country)
higher_earn_country=total_country["United-States"]/total_person*100
print(f'{round(higher_earn_country,2)}%')
```
